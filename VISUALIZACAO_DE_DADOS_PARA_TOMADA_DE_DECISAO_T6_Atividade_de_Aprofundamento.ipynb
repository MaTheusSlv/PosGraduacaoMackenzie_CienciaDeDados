```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#XLSX original gerado no Site do IBGE
#xlsx = pd.read_excel('https://drive.google.com/uc?id=1txKEao_Sf_T9jV2JVPjII_Ac_vDi5NGI')

#csv formatado
df = pd.read_csv('https://drive.google.com/uc?id=1wqwgFEhj7nCqdoWFZJ3Pvnd8KT3oC9KU', delimiter=';')

df
```

```python
#Análise exploratória
print('Qtd. Linhas , Qtd, de Colunas:')
print(np.shape(df))
print('\nTipos de dados:')
print(df.dtypes)
print(f'\nQuantidade de nulls:')
print(df.isnull().sum())
df.describe()
```

```python
#Ignorando os avisos do Python
import warnings
pd.options.mode.chained_assignment = None
warnings.filterwarnings("ignore", category=UserWarning, message="FixedFormatter should only be used together with FixedLocator")

#renomeando as colunas
df.columns = ['faixa_etaria','local','Janeiro','Fevereiro','Março','Abril','Maio','Junho','Julho','Agosto','Setembro','Outubro','Novembro','Dezembro']

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

#visualização 1
x = df.drop(columns=['faixa_etaria','local']).columns #meses
y = df.drop(columns=['faixa_etaria','local']).sum() #quantidades

ax1.bar(x, y)

ax1.set_title('Quantidade de óbitos mensais em 2022')
ax1.set_xlabel('Meses')
ax1.set_ylabel('Quantidade')
ax1.set_xticklabels(x, rotation=45)

#visualização 2
x2 = df['local'].sort_values().unique() #locais
y2 = df.drop(columns=['faixa_etaria']).groupby('local').sum().sum(axis=1) #quantidades

ax2.bar(x2, y2)

ax2.set_title('Quantidade de óbitos por local em 2022')
ax2.set_xlabel('Locais')
ax2.set_ylabel('Quantidade')
ax2.set_xticklabels(x2, rotation=45)
ax2.ticklabel_format(style='plain', axis='y')

plt.tight_layout()
plt.show()
```

**Resultado da Análise do Dataset**

Analisando o conjunto de dados foi possível constatar que ele é composto por 16 registros(linhas) e 14 variáveis(colunas) onde se encontram os valores totais de óbitos registrados no ano de 2022, sendo que em nenhum dos registros há a presença de valores nulos e, das variáveis presentes duas são qualitativas, uma ordinal (Idade do(a) falecido(a)) e outra nominal (Local de ocorrência), as demais variáveis (Janeiro, Fevereiro, Março, Abril, Maio, Junho, Julho, Agosto, Setembro, Outubro, Novembro, Dezembro) são quantitativas discretas.
Ao analisar a plotagem dos gráficos de barras ainda para conhecer o dataset, pode-se observar que o maior número de óbitos registrados se concentrava no mês de Janeiro, o número então decaia até Abril e recomeçava a crescer atingindo um novo pico em Julho, porém ainda menor que o de Janeiro, e volta a decair até Setembro, já nos meses finais do ano os números parecem se manter próximos até a chegada de Dezembro onde volta a crescer um pouco. Outro ponto observado é de que há um grande número de óbitos registrados em Hospitais, talvez por conta dos processos corriqueiros de registros que ocorrem nesses hospitais ou até mesmo em decorência da natureza dos momentos pré-óbito onde a pessoa é naturalmente levada aos hospitais com a esperança de serem reanimadas.

```python
#ESTUDO
df = pd.read_csv('https://drive.google.com/uc?id=1wqwgFEhj7nCqdoWFZJ3Pvnd8KT3oC9KU', delimiter=';')

#renomeando as colunas
df.columns = ['faixa_etaria','local','Janeiro','Fevereiro','Março','Abril','Maio','Junho','Julho','Agosto','Setembro','Outubro','Novembro','Dezembro']

#dados do gráfico 1

#trimestre1 sem hospital
trimestre1 = df[['faixa_etaria','local','Janeiro','Fevereiro','Março']]
trimestre1['totaltrimestre1'] = trimestre1.drop(columns=['faixa_etaria','local']).sum(axis=1)
trimestre1 = trimestre1.loc[trimestre1['local'] != 'Hospital']
#trimestre2 sem hospital
trimestre2 = df[['faixa_etaria','local','Abril','Maio','Junho']]
trimestre2['totaltrimestre2'] = trimestre2.drop(columns=['faixa_etaria','local']).sum(axis=1)
trimestre2 = trimestre2.loc[trimestre2['local'] != 'Hospital']
#trimestre3 sem hospital
trimestre3 = df[['faixa_etaria','local','Julho','Agosto','Setembro']]
trimestre3['totaltrimestre3'] = trimestre3.drop(columns=['faixa_etaria','local']).sum(axis=1)
trimestre3 = trimestre3.loc[trimestre3['local'] != 'Hospital']
#trimestre4 sem hospital
trimestre4 = df[['faixa_etaria','local','Outubro','Novembro','Dezembro']]
trimestre4['totaltrimestre4'] = trimestre4.drop(columns=['faixa_etaria','local']).sum(axis=1)
trimestre4 = trimestre4.loc[trimestre4['local'] != 'Hospital']

#totais por trimestre sem hospital
trimestres = pd.concat([df[['faixa_etaria','local']].loc[df['local'] != 'Hospital'], trimestre1['totaltrimestre1'], trimestre2['totaltrimestre2'], trimestre3['totaltrimestre3'], trimestre4['totaltrimestre4']], axis=1)


fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

#gráfico 1 - Número de óbitos trimestrais por local
x = trimestres.drop(columns=['faixa_etaria','local']).columns #trimestres
y = trimestres.drop(columns='faixa_etaria').groupby('local').sum() #quantidade

tam_barra = 0.3
posicao = range(len(x))

for i, local in enumerate(y.index):
    ax1.bar([p + tam_barra * i for p in posicao], y.loc[local], width=tam_barra, label=local)

ax1.set_title('Número de óbitos trimestrais\nque ocorreram fora de Hospitais\n(Todas as idades)')
ax1.set_xlabel('Trimestres')
ax1.set_ylabel('Quantidade')
ax1.legend(title='Local onde ocorreu')
ax1.set_xticks([p + tam_barra * (len(y.index) / 2) for p in posicao])
ax1.set_xticklabels(x, rotation=45)

#dados do gráfico 2

#faixas etárias sensíveis sem hospital
sensivel = df.loc[df['faixa_etaria'] != '15 a 84 anos']
sensivel = sensivel.loc[sensivel['local'] != 'Hospital']

#gráfico 2 - Número de óbitos mensais das faixa etárias mais sensíveis por local
x2 = sensivel.drop(columns=['faixa_etaria','local']).columns #meses
y2 = sensivel.drop(columns='faixa_etaria').groupby('local').sum() #quantidades

for local in y2.index:
    ax2.bar(x2, y2.loc[local], label=local)

ax2.set_title('Número de óbitos mensais das faixas\netárias mais sensíveis fora dos Hospitais\n(Menos de 1, 1 a 14 e 85 anos ou mais)')
ax2.set_xlabel('Meses')
ax2.set_ylabel('Quantidade')
ax2.legend(title='Local onde ocorreu')
ax2.set_xticklabels(x2, rotation=45)

plt.tight_layout()
plt.show()
```
