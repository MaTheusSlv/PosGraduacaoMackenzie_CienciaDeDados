```python
import scrapy
import os
import pathlib
import csv

class MinhaSpider(scrapy.Spider):
  name='MeuPrimeiroRobo'

  #abre arquivo local com os links
  start_urls = [f"{pathlib.Path(os.path.abspath('/content/robo_mat/robo_mat/spiders/F1 News - Motorsport.html')).as_uri()}"]

  #apaga o csv se já esxistir
  if os.path.exists("MeuPrimeiroRobo.csv"):
    os.remove("MeuPrimeiroRobo.csv")

  #função que interpreta a página com as notícias
  def parse(self, response):
    for link in response.css(".ms-grid-hor-m").css("a::attr(href)").getall():
      text_page= f"https://www.motorsport.com{link}"
      yield scrapy.Request(text_page, callback=self.parse_text)

  #função que interpreta as páginas com os textos
  def parse_text(self, response):
    content=""
    for line in response.css('div.ms-article-content p::text').getall():
      content = content +"".join(line)+" "

    #monta estrutura e recebe o conteúdo do arquivo
    post = {
        'Título da Notícia':response.css('div.ms-entity-detail-header h1.text-h1::text').get(),
        'Conteúdo da Notícia':content
    }

    #escreve o CSV
    file_exists = os.path.isfile("MeuPrimeiroRobo.csv")
    with open('MeuPrimeiroRobo.csv','a',newline='',encoding="utf-8") as output_file:
      dict_writer = csv.DictWriter(output_file, ['Título da Notícia','Conteúdo da Notícia'])
      if not file_exists:
        dict_writer.writeheader()
      dict_writer.writerows([post])
```
