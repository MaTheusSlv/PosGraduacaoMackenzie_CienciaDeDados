# Seu Código aqui.

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.title('Relação Masculina da Performance x Idade', weight='bold')
sns.barplot(y= data.performance, x= data.age[(data.gender == 'Male') & (data.age >= 30) & (data.age <= 45)])
plt.hlines(data.performance.mean(), -8, 16, linestyles=':')
plt.text(-8, data.performance.mean()+0.1, 'Média Geral')
plt.ylabel('Performance')
plt.xlabel('Idade')

plt.subplot(1,2,2)
plt.title('Relação Feminina da Performance x Idade', weight='bold')
sns.barplot(y= data.performance, x= data.age[(data.gender == 'Female') & (data.age >= 30) & (data.age <= 45)])
plt.hlines(data.performance.mean(), -8, 16, linestyles=':')
plt.text(-8, data.performance.mean()+0.1, 'Média Geral')
plt.ylabel('Performance')
plt.xlabel('Idade')

plt.tight_layout()
plt.show()
```

Nos gráficos acima pode-se avaliar a relação entre a pontuação de desempenho dos colaboradores pela suas respectivas idades, sendo divididas por gênero Masculino e Feminino onde por esta avaliação foi possivel perceber um fato curioso, em que a performance Masculina, ao se aproximar dos 40 anos de idade, ela tende a diminuir gradativamente e ficar abaixo da média geral, fato que não ocorre com a amostra Feminina, onde a mesma mantém um certo padrão próximo da média geral de performance quando se aproxima da mesma faixa de 40 anos. Este fato observado pode ser um ponto de partida para entender o porque disso ocorrer, como um possível reflexo da falta de confiança que afeta o desempenho profissional dos homens por algum motivo psicológico por alcançar esta idade, ou até pode-se pressupor que nesta idade seja mais difícil a adaptação à novas tecnologias/processos por parte do colaborador, demandando uma atenção especial da organização em ambos os casos.

```python
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.title('Relação Performance x Cargos Relacionados à TI', weight='bold')
sns.barplot(y= data.performance, x= data.jobtitle[(data.jobtitle == 'Data Scientist') | (data.jobtitle == 'IT') | (data.jobtitle == 'Software Engineer')])
plt.hlines(data.performance.mean(), -8, 16, linestyles=':')
plt.text(-8, data.performance.mean()+0.1, 'Média Geral')
plt.xticks(rotation=90)
plt.ylabel('Performance')
plt.xlabel('Cargo')

plt.subplot(1,2,2)
plt.title('Relação Performance x Cargo Não Relacionados à TI', weight='bold')
sns.barplot(y= data.performance, x= data.jobtitle[(data.jobtitle == 'Driver') | (data.jobtitle == 'Financial Analyst') | (data.jobtitle == 'Graphic Designer')| (data.jobtitle == 'Manager') | (data.jobtitle == 'Marketing Associate')| (data.jobtitle == 'Sales Associate') | (data.jobtitle == 'Warehouse Associate')])
plt.hlines(data.performance.mean(), -2, 7, linestyles=':')
plt.text(-2, data.performance.mean()+0.1, 'Média Geral')
plt.xticks(rotation=90)
plt.ylabel('Performance')
plt.xlabel('Cargo')

plt.tight_layout()
plt.show()
```

Os dois gráficos sobre a relação da performance com cargos relacionados à TI e com cargos não relacionados permite que o gestor possa avaliar o desempenho dos colaboradores que têm esses cargos da área de TI e analisar se estes estão próximos da média geral de performance ou se estão acima dela, que no caso a maioria está com uma performance acima, no caso de estarem abaixo, o mesmo pode a partir da observação tomar medidas para otimizar a performance desses cargos no caso do primeiro gráfico. Já no segundo pode-se observar o desempenho de outros cargos em relação à média geral, permitindo ao gestor de TI por exemplo, entender se os cargos que estão sob a sua supervisão está com um rendimento satisfatório em relação às outras áreas da empresa, sendo o gráfico uma ferramenta de comparação de lideranças importante.
