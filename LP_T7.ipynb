# Seu Código aqui.

```r
library('ggplot2')
library('tidyverse')

ggplot(filter(data, Country == 'United States of America'), aes(x=Year, y=infant.deaths)) +
  geom_point() + geom_smooth() +
  ggtitle('Mortalidade Infantil nos EUA', subtitle = 'De 2000 a 2015') + theme_linedraw() +
  xlab('Ano') +
  ylab('Nível de Mortalidade')
```

No gráfico gerado sobre a mortalidade infantil nos Estados Unidos, pode-se perceber que no período de tempo disponibilizado da amostra entre os anos de 2000 a 2015, a mortalidade no País teve uma grande tendência à diminuição podendo representar que o País neste período pôde estar alocando corretamente os seus recursos para a área da saúde, impedindo assim que essa taxa de mortalidade se mantenha numa crescente ou constante, apresentando diminuição no número de mortes de crianças.

```r
ggplot(filter(data, Country == 'Brazil' & Year >= 2005), aes(x=Year, y=Schooling)) +
  geom_point() + geom_smooth() +
  ggtitle('Escolaridade no Brasil', subtitle = 'De 2005 a 2015') + theme_gray() +
  xlab('Ano') +
  ylab('Nível de Escolaridade')
```

Avaliando o nível de escolaridade no Brasil no período de 10 anos entre 2005 e 2015 no gráfico, é possível ter a grata visualização de que o nosso País apresentou melhora nos níveis escolares dentro desse período, demonstrando um possível interesse maior dos alunos e/ou uma oferta maior das possibilidades de se estudar sendo ofertadas ou proporcionadas pelos governos do período ou até pela situação econômica do País que possibilitaria uma maior oferta de empregos e consequentemente recursos para as pessoas buscarem a capacitação.
