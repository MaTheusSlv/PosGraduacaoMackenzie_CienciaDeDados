# Construa um contador de palavras

Textos são uma fonte importante de informações para Ciência de Dados. Eles aparecem nos posts das redes sociais, na avaliação de serviços e produtos etc. Contar a frequência (ou a quantidade) que esses termos aparecem é a base de muitos sistemas de recomendação, análise de sentimento etc.

Construa um contador de palavras de um texto que, dado um texto, produz um dicionário `contagem` como a palavra sendo chave e suas quantidades o valor da chave. Veja o exemplo:

Entrada do programa:

```
texto = '''No meio do caminho tinha uma pedra
tinha uma pedra no meio do caminho
tinha uma pedra no meio do caminho tinha uma pedra.
'''
```

Saída:

```
{'no': 3, 'meio': 3, 'do': 3, 'caminho': 3, 'tinha': 4, 'uma': 4, 'pedra': 4}
```

````python
data = '''Lorem ipsum dolor sit amet consectetur adipiscing elit per blandit,
 arcu nullam platea condimentum hendrerit placerat mauris.
 Enim montes per suscipit semper dui porttitor et sapien praesent aptent,
 ac sit natoque sollicitudin tristique vitae laoreet viverra euismod.
 In eget phasellus libero volutpat integer dictumst nisl venenatis tellus parturient ac laoreet,
 feugiat cras nunc semper sapien vulputate mus massa lectus consequat morbi habitant porttitor,
 donec auctor vel risus pharetra pulvinar lobortis quisque bibendum scelerisque tincidunt.
 Fames mollis tincidunt aenean platea ullamcorper lobortis non sodales litora,
 interdum amet viverra maximus congue urna malesuada dis, condimentum dolor iaculis hac dui nisl habitant elementum.
 Nibh odio aptent maximus parturient natoque sodales nulla est lorem,
  mollis quam dolor quis tincidunt id consequat penatibus amet donec,
  condimentum inceptos pulvinar aliquam iaculis justo litora consectetur.
  Etiam viverra quam cursus orci donec libero ligula in hac, lacus euismod tincidunt semper pretium a rhoncus facilisi suspendisse,
  aptent porttitor dui eleifend pharetra duis facilisis integer.
  Ipsum efficitur ornare himenaeos tincidunt mollis adipiscing lacinia nascetur volutpat et blandit elit,
  habitant felis arcu integer pharetra ligula posuere pellentesque ullamcorper donec ac.
  Blandit mollis turpis molestie non porta sodales laoreet, platea hendrerit mi senectus morbi nunc ligula mattis,
   aptent auctor litora etiam facilisis euismod. Accumsan torquent dolor rhoncus odio elit vivamus quam lacinia tempus praesent dui maecenas netus congue,
   class ex enim mauris quisque ornare posuere hac iaculis tellus parturient ac vel.
'''

# Remove caracteres especiais e transforma todas as letras em minúsculas
data = data.lower().replace(".", "").replace(",", "").replace(":", "").replace(";", "").replace("!", "").replace("?", "").replace("-", "")

# Separa o texto em uma lista de palavras
lista_palavras = data.split()

print('Lista... \n')
print(lista_palavras)
print()

# Cria um dicionário para armazenar a contagem de cada palavra
contagem = {}

#
# CRIE O DICIONÁRIO CONTAGEM AQUI ----------------------------------------------
#
for i in lista_palavras:
    if i in contagem:
        contagem[i] += 1
    else:
        contagem[i] = 1

#
#-------------------------------------------------------------------------------
#

# Imprime a contagem de cada palavra
print('contagem... \n')
print(contagem)
````
